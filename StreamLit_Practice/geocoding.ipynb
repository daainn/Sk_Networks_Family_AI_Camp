```python
from geopy.geocoders import Nominatim
import pandas as pd


geolocator = Nominatim(user_agent="myGeocoder")

areas = [
    'Darebin', 'Moyne', 'Baw Baw', 'Mitchell', 'Wyndham', 'Casey', 'Hepburn',
    'Greater Geelong', 'Hindmarsh', 'Northern Grampians', 'Mildura', 'Mount Alexander',
    'Golden Plains', 'Frankston', 'Indigo', 'Mornington Peninsula', 'Total', 'Kingston',
    'Gannawarra', 'Colac-Otway', 'Murrindindi', 'Warrnambool', 'Greater Bendigo', 'West Wimmera',
    'Central Goldfields', 'Moira', 'Latrobe', 'Greater Shepparton', 'Boroondara', 'Knox',
    'East Gippsland', 'Nillumbik', 'Wangaratta', 'Yarra', 'Buloke', 'Manningham', 'Whittlesea',
    'Hobsons Bay', 'Hume', 'Cardinia', 'Yarriambiack', 'Port Phillip', 'Queenscliffe', 'Glenelg',
    'Bayside', 'South Gippsland', 'Moreland', 'Whitehorse', 'Surf Coast', 'Maribyrnong', 'Strathbogie',
    'Maroondah', 'Justice Institutions and Immigration Facilities', 'Campaspe', 'Ararat', 'Banyule',
    'Monash', 'Benalla', 'Macedon Ranges', 'Pyrenees', 'Alpine', 'Stonnington', 'Greater Dandenong',
    'Glen Eira', 'Unincorporated Vic', 'Brimbank', 'Corangamite', 'Bass Coast', 'Southern Grampians',
    'Loddon', 'Melbourne', 'Ballarat', 'Wellington', 'Swan Hill', 'Horsham', 'Melton', 'Moonee Valley',
    'Moorabool', 'Mansfield', 'Towong', 'Wodonga', 'Yarra Ranges'
]


geocode_results = []


for area in areas:
    location = geolocator.geocode(area + ", Australia")
    if location:
        geocode_results.append({'area': area, 'lat': location.latitude, 'lng': location.longitude})
    else:
        geocode_results.append({'area': area, 'lat': None, 'lng': None})


df_geocode = pd.DataFrame(geocode_results)
```

```python
df_geocode.rename(columns={"area": "Local Government Area"}, inplace=True)

df_geocode["Local Government Area"] = df_geocode["Local Government Area"].str.strip().str.lower()
df_crime["Local Government Area"] = df_crime["Local Government Area"].str.strip().str.lower()

merged_df = pd.merge(df_geocode, df_crime, on="Local Government Area", how="inner")
```

```python
merged_df["Incidents Recorded"] = merged_df["Incidents Recorded"].str.replace(",", "")

merged_df["Criminial Incident Rate"] = merged_df["Criminial Incident Rate"].str.replace(",", "")
merged_df["Criminial Incident Rate"] = merged_df["Criminial Incident Rate"].str.split(".").str[0]

merged_df["Rate per 100,000 population"] = merged_df["Rate per 100,000 population"].str.replace(",", "")
merged_df["Rate per 100,000 population"] = merged_df["Rate per 100,000 population"].str.split(".").str[0]

merged_df["Criminial Incident Rate"] = pd.to_numeric(merged_df["Criminial Incident Rate"], errors="coerce")
merged_df["Incidents Recorded"] = pd.to_numeric(merged_df["Incidents Recorded"], errors="coerce")
merged_df["Rate per 100,000 population"] = pd.to_numeric(merged_df["Rate per 100,000 population"], errors="coerce")
average_num = ["Criminial Incident Rate", "Incidents Recorded", "Rate per 100,000 population", "lng", "lat"]
average_df = merged_df.groupby("Local Government Area")[average_num].mean().reset_index()
```

```python
import pydeck as pdk

chart_data = pd.read_csv("crime_geocode.csv")

chart_data_cleaned = chart_data.dropna(subset=['lat', 'lng', 'Criminial Incident Rate'])

r = pdk.Deck(
    map_style=None,
    initial_view_state=pdk.ViewState(
        latitude=chart_data['lat'].mean(),
        longitude=chart_data['lng'].mean(),
        zoom=6,
        pitch=50,
    ),
    layers=[
        pdk.Layer(
            "HexagonLayer",
            data=chart_data,
            get_position='[lng, lat]',
            radius=2000,  # 헥사곤 반지름 (미터 단위)
            elevation_scale=0.2,  # 높이 스케일
            elevation_range=[0, 4000],  # 높이 범위
            extruded=True,  # 3D 효과
            pickable=True,
            coverage=0.8,  # 헥사곤 내 데이터 커버리지
            get_elevation_value="Criminial Incident Rate",  # 높이 기준 컬럼
            get_fill_color="[Criminial Incident Rate * 255 / 17366.1, 50, 100]",
            ),
        ],
    )

r.show()
```
