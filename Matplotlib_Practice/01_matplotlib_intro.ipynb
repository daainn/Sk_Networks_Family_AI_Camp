### figure | axes

* figure : 전체 그림이 그려지는 컨테이너
* axes : 그래프가 실제로 그려지는 영역 (좌표, 데이터)

```python
fig = plt.figure(figsize=(6,4), facecolor='pink')
ax = plt.gca()
ax.plot([1,2,3], [4,5,6])
print(ax)

plt.title('Hello Matplotlib')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### plot 메서드의 매개변수

```python
plt.figure(figsize=(6,4))
plt.plot(
    x,
    y,
    color='black',
    marker='o',
    markersize = 8,
    linestyle='--',
    linewidth = 2
)
plt.show()
```

```python
plt.plot(x,y)
plt.title('제목')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show()
```

* ticks

```python
plt.figure(figsize = (3,2))

x = np.arange(1, 101)
y = np.linspace(2,200,100)

plt.xticks(ticks=np.arange(0,101,5), rotation = 90)
plt.yticks(rotation=45)

plt.plot(x, y)
plt.show()
```

* lim : 보여지는 범위 재지정

```python
plt.figure(figsize = (3,2))

x = np.arange(1, 101)
y = np.linspace(2,200,100)

plt.xlim(1,10)
plt.ylim(2,10)

plt.plot(x, y)
plt.show()
```

* legend

```python
plt.figure(figsize = (3,2))

x = np.arange(1, 101)
y = np.linspace(2,200,100)
y2 = np.linspace(3, 300, 100)


plt.plot(x, y, label='시간')
plt.plot(x, y2, label='공간')
# loc 범례 위치 (기본값: best 최적의 위치)
# - (upper | lower) (left | right)
# - center
plt.legend(loc = 'lower right')

plt.show()
```

### 그래프 여러 개 그리기

```python
plt.figure(figsize=(3,2))

x = np.arange(1, 11)
y = x * 3
y2 = x * 2


plt.plot(x, y, label='선', color = 'red', marker='o')
plt.bar(x, y2, label='바', color = 'green')

plt.legend()
plt.grid(True)

plt.show()
```

```python
# subplot 이용 -> 여러 그래프를 각각 그리기
fig, ax = plt.subplots(1,2, figsize=(10, 5))

x1 = np.arange(1, 11)
y1 = x1 * 2
x2 = np.arange(11, 20)
y2 = x2 * 2

ax[0].plot(x1, y1, label = '선', color = 'red')
ax[0].set_xlabel('X1')
ax[0].set_ylabel('Y1')
ax[0].set_title('빨간 선')
ax[0].legend()
ax[0].grid()

ax[1].bar(x2, y2, label = '바', color = 'pink')
ax[1].set_xlabel('X2')
ax[1].set_ylabel('Y2')
ax[1].set_title('빨간 선')
ax[1].legend()
ax[1].grid()
```
